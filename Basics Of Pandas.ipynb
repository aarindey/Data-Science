```python
# Finding the top 5 batsman who hit maximum number of 4s
mask=delivery['batsman_runs']==4
only4s=delivery[mask]
top4s=only4s.groupby('batsman')
top4s['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
top4s['batsman_runs'].count().sort_values(ascending=False).head()
#Number of 4s
```

```python
# .size() method can also be used 
top4s.size().sort_values(ascending=False).head()
```

```python
# Which are the three teams against whom V Kohli has scored the maximum runs
mask2=delivery['batsman']=='V Kohli'
delivery[mask2].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# Creating a function which returns the name of the team as per above logic
def maxrunsagainst(batsman_name):
    mask2=delivery['batsman']==batsman_name
    return delivery[mask2].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
maxrunsagainst('V Kohli')
```

```python
maxrunsagainst('DA Warner')
```

```python
maxrunsagainst('SK Raina')
```

```python
# The Merge function isin() function
# Find the most destructive death over batsman in the history of IPL
# Strike Rate=(Number of runs/Number of balls)  *100
# Min batsman 200 balls in over 16-20
death_over=delivery[delivery['over']>=16]
# we need balls played 
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
#Runs scored by all these 43 batsman 
# Balls played by these 43 batsman

#we are concerned about overs 16-20 only
final=death_over[death_over['batsman'].isin(batsman_list)]

runs=final.groupby('batsman')['batsman_runs'].sum()
balls_played=final.groupby('batsman')['batsman_runs'].count()
sr=(runs/balls_played)*100
sr
```
