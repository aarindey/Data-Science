```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset('tips')
plt.style.use('fivethirtyeight')
```

```python
data.head()
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',hue='smoker',data=data)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',hue='smoker',style='sex',data=data)
plt.show()
```

```python
# hue and style holds categorical data
# size holds numeical data
sns.relplot(x='total_bill',y='tip',kind='scatter',hue='smoker',style='sex',size='size',sizes=(15,200),data=data)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
plt.show()
#scatterplot() can also be used
```

```python
# Categorical Data
# catplot
# stripplot
# swarmplot
# by default stripplot
sns.catplot(x='day',y='tip',data=data)
plt.show()
# jitter parameter deals with the noise separating each point with the other
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data)
plt.show()
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=data)
plt.show()
# no style and size parameters
```

```python
sns.swarmplot(x='day',y='tip',hue='sex',data=data)
plt.show()
```

```python
# boxplot

#univariate
sns.boxplot(data['tip'])
plt.show()
```

```python
# boxplot
sns.boxplot(data['total_bill'])
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=data)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=data)
plt.show()
```

```python
#Violin plots
sns.violinplot(data['total_bill'])
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data)
plt.show()
```

```python
sns.catplot(x='day',hue='sex',y='total_bill',kind='violin',data=data)
plt.show()
```

```python
sns.catplot(x='day',hue='sex',y='total_bill',kind='violin',split=True,data=data)
plt.show()
```

```python
# Bar Plots and Count Plots
sns.catplot(x='smoker',y='total_bill',kind='bar',data=data)
plt.show()
```

```python
# Bar Plots and Count Plots
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=data)
plt.show()
```

```python
sns.catplot(x='sex',kind='count',data=data)
plt.show()
```

```python
# Heatmap
# 2 categorical data or a discrete numerical data
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x,cbar=False)
plt.show()
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt=)
plt.show()
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='viridis')
# cmap ='summer'/ 'winner'
plt.show()
```

```python
#Clustermap
# plot heirarchiacally clustered rows and cols
sns.clustermap(x)
#metric=eclidean by default

plt.show()
```

```python
sns.clustermap(x,metric='correlation')
#metric=eclidean by default

plt.show()
```

```python
sns.clustermap(x,metric='correlation',z_score=0,annot=True)
#metric=eclidean by default

plt.show()
```

```python
sns.clustermap(x,metric='correlation',row_cluster=False,z_score=0,annot=True)
#metric=eclidean by default

plt.show()
```

```python
# Distplots 
# combines 3 things in a graph
# Histogram
# Kernel Density Estimation (PDF)
# Rugplot
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
plt.show()
```

```python
sns.distplot(titanic['age'],kde=False)
plt.show()
```

```python
sns.distplot(titanic['age'],hist=False)
plt.show()
```

```python
sns.distplot(titanic['age'],kde=False,hist=False,rug=True)
plt.show()
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
plt.show()
```
